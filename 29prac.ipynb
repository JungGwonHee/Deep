```python
import matplotlib.pyplot as plt
import numpy as np

bream_length = [25.4,26.3,26.5,29.0,29.0,29.7,29.7,30.0,30.0,30.7,31.0,31.0,31.5,32.0,32.0,32.0,33.0,33.0,33.5,33.5,34.0,34.0,34.5,35.0,35.0,35.0,35.0,36.0,36.0,37.0,38.5,38.5,39.5,41.0,41.0]

bream_weight = [242.0,290.0,340.0,363.0,430.0,450.0,500.0,390.0,450.0,500.0,475.0,500.0,500.0,340.0,600.0,600.0,700.0,700.0,610.0,650.0,575.0,685.0,620.0,680.0,700.0,725.0,720.0,714.0,850.0,1000.0,920.0,955.0,925.0,975.0,950.0]

smelt_length = [9.8,10.5,10.6,11.0,11.2,11.3,11.8,11.8,12.0,12.2,12.4,13.0,14.3,15.0]

smelt_weight = [6.7,7.5,7.0,9.7,9.8,8.7,10.0,9.9,9.8,12.2,13.4,12.2,19.7,19.9]

bream_data = [[l,w] for l, w in zip(bream_length, bream_weight)] #list의 list를 scikit이 잘 알아보도록 만들어 줌
smelt_data =  [[l,w] for l, w in zip(smelt_length, smelt_weight)] #zip 함수로 길이와 무게를 같은 shape으로 정렬시킴

fish_data = bream_data + smelt_data
fish_target = [1]*35 + [0]*14 # 정답을 만들어 주는것임, 정답이 1 즉 찾고 싶은 데이터가 1이 되어야함

ar_fish_data = np.array(fish_data)
ar_fish_target = np.array(fish_target)
display(ar_fish_data, ar_fish_data.shape)
display(ar_fish_target, ar_fish_target.shape)

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

#k-최근접 이웃
#분류가 되지 않은 데이터를 근접한 값과 같도록 분류하는 방법
from sklearn.neighbors import KNeighborsClassifier#scikit_learn
# kn = KNeighborsClassifier() #import as kn과 비슷한 내용임
# kn.fit(fish_data, fish_target)
# kn.score(fish_data, fish_target)
# kn.predict([[30,600]]) #새로운 생선 예측
model = KNeighborsClassifier(n_neighbors=3)

model.fit(ar_fish_data, ar_fish_target) 
print(model.score(ar_fish_data, ar_fish_target))
model.predict([[30,300]])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
#target이 정답값
dir(iris)
# sepal(꽃받침) length,sepal width, petal(꽃잎) length, petal width
iris.feature_names
iris.target_names  # setosa, versicolor, virginica >> iris의 종류인듯

display(iris.target, iris.target.shape)  # target의 값과 모양
display(iris.target[iris.target == 0].shape, iris.target[iris.target ==
        1].shape, iris.target[iris.target == 2].shape)  # 0,1,2의 target
display(iris.data.shape, iris.data[:5])  # slice

#머신러닝은 훈련과정과 테스트 과정이 있다
#훈련 세트를 가지고 예측 모델을 훈련시킨 후 테스트세트로 훈련 성과 판단

#train_test_split(arrays, test_size(default = 0.25),train_size(default = test_size의 나머지), random_state(데이터 셔플시 시드값), shuffle(셔플여부설정 default = true), stratify(지정한 data의 비율유지))
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target)  # 데이터 분할, train:test = 8:2로 나눠서 학습하기 위해
#전통적으로 데이터는 대문자X 타겟(또는 레이블)은 소문자
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
display(X_train[:5], y_train[:5])

# dataframe안에 훈련 세트와 칼럼이름넣기
iris_df = pd.DataFrame(X_train, columns=iris.feature_names)

iris_df[:5]

pd.plotting.scatter_matrix(iris_df, c=y_train, s=60,
                           alpha=0.8, figsize=[12, 12])
#scatter_matrix는 pandas에서 이용, 데이터에서 2개식 짝지어서 6개의 산점도
print('')


model = KNeighborsClassifier(n_neighbors=1)  # 기본값은 5,주변 이웃값을 정함
model.fit(X_train, y_train)  # fit 데이터를 줄테니 모델을 만들어봐 >> 이를 이용하여 예측

display(model.score(X_test, y_test))

model.predict([[6, 3, 4, 1.5]])  # 임의의 점을 주고 예측하는 것
#꽃받침 길이, 꽃받침 너비, 꽃잎 길이, 꽃잎 너비 입력하여 데이터세트에 대한 값중 종류를 뽑아냄
```
